## Image pre-processing

```python
# Read in the image
image = mpimg.imread('images/car.jpeg')

# Print out the image dimensions
print('Original Image dimensions:', image.shape)

plt.imshow(image);
```

#### Convert color image to gray scale - This helps in dimentionality reduction

```python
# Change from color to grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
print('Grayscale Image dimensions:', gray_image.shape)

plt.imshow(gray_image, cmap='gray');
```

Now the image has become 2 dimensional. 

```python
gray_image.shape
```

There will be a total of 427*640 pixels, and each pixel will have a number assigned to it. Therefore, each pixel becomes the input feature. for e.g. let's say if you had a 5x5 pixel image, you would have a total of 25 pixels and those 25 pixel values are the input features to the CNN network. See image below:

![alt text](images/pixels.png "Title")

This process of going from 2D to 1D vector is known as **flattening**. Because these networks cannot take 2D data as input.

```python
# Prints specific grayscale pixel values
# What is the pixel value at x = 400 and y = 300 (on the body of the car)?

x = 400
y = 300

print(gray_image[y,x])
```

```python
# Finds the maximum and minimum grayscale values in this image

max_val = np.amax(gray_image)
min_val = np.amin(gray_image)

print('Max: ', max_val)
print('Min: ', min_val)
```

```python
# Creating a 5x5 image using just grayscale, numerical values
tiny_image = np.array([[0, 20, 30, 150, 120],
                      [200, 200, 250, 70, 3],
                      [50, 180, 85, 40, 90],
                      [240, 100, 50, 255, 10],
                      [30, 0, 75, 190, 220]])

# To show the pixel grid, use matshow
plt.imshow(tiny_image, cmap='gray')
```
